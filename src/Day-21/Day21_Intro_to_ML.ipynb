Machine Learning is a way of teaching computers to learn patterns from data instead of explicitly programming every rule. In traditional programming, we provide the input data and define the rules to get an output, but in machine learning, we provide the data along with the correct answers, and the system learns the rules by itself. This allows computers to make predictions or decisions when new data is given, even if the exact situation was not programmed before.

Supervised Learning

Supervised learning is a type of machine learning where the model is trained using data that already includes the correct answers. The system learns by comparing its predictions with the actual answers and improves over time. It is mainly used for prediction and classification tasks.

Unsupervised Learning

Unsupervised learning works with data that does not have any predefined labels or answers. The model tries to explore the data and identify hidden patterns, structures, or groups on its own. It is commonly used when we don’t know the outcome beforehand.

Reinforcement Learning

Reinforcement learning is based on learning through experience. The model interacts with an environment and receives rewards for correct actions and penalties for wrong ones. Over time, it learns the best strategy to maximize rewards.

| Learning Type        | How it Works                                      | Examples                          |
|---------------------|--------------------------------------------------|----------------------------------|
| Supervised Learning | Learns from labeled data (input + correct output) | Email spam detection, Predicting house prices |
| Unsupervised Learning | Finds patterns in unlabeled data               | Customer segmentation, Grouping similar products |
| Reinforcement Learning | Learns through rewards and penalties          | Self-driving cars, Playing video games |
